```python
# set-2 q3

mean = 38
std = 6
n = 1000

# defining X1 and X2

x1 = np.random.normal(mean, std, n)
x2 = np.random.normal(mean, std, n)

# for 2 * x1
x1_2 = x1 * 2

# for x1 + x2
x1_x2 = x1 + x2
```

```python
plt.figure(figsize=(12,8))

plt.subplot(221)
sns.kdeplot(x1, bw_adjust=1.5)
plt.title("X1")


plt.subplot(222)
sns.kdeplot(x2, bw_adjust=1.5)
plt.title("X2")


plt.subplot(223)
sns.kdeplot(x1_2, bw_adjust=1.5)
plt.title("2 * X1")


plt.subplot(224)
sns.kdeplot(x1_x2, bw_adjust=1.5)
plt.title("X1 + X2")

plt.subplots_adjust(wspace=.4, hspace=0.5)
plt.show()
```

```python
print(f"Mean of X1 is {np.round(x1.mean(),2)},std of X1 is {np.round(x1.std(),2)}")
print(f"Mean of X2 is {np.round(x2.mean(),2)},std of X2 is {np.round(x2.std(),2)}")
print(f"Mean of 2*X1 is {np.round(x1_2.mean(),2)},std of 2*x1 is {np.round(x1_2.std(),2)}")
print(f"Mean of X1+X2 is {np.round(x1_x2.mean(),2)},std of X1 + X2 is {np.round(x1_x2.std(),2)}")
```

```python
# q5

# profit1 division 
m1 =5
var1 = 3**2

# profit2 division 
m2=7
var2 = 4**2

#profits are in $ we need to convert it to ind, assuming 1$=Rs. 45
ind = 45 

total_mean = (m1+m2) * ind

total_var = (var1 + var2) 
std_ = total_var ** 0.5
total_std = std_ * ind 

# a.
ci = st.norm.interval(0.95,total_mean, total_std)
print("95% confidence interval",np.round(ci,2))

# b.

per_5th = st.norm.ppf(0.05, total_mean, total_std)
print("5th percentile of profit for the company is ", np.round(per_5th))

# c.
div1 = st.norm.cdf(0,m1,var1**2)
div2 = st.norm.cdf(0,m2,var2**2)
print(div1, div2)
# division 2 has more chance of making loss in a given year.
```
